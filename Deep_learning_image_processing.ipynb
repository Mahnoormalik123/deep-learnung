```python
from PIL import Image
from IPython.display import display

image_path = "demo1.jpg.jpeg"
image_file = Image.open(image_path)

image_file.save("demo1_quality95.jpg", quality=5000)
image_file.save("demo1_quality25.jpg", quality=25)
image_file.save("demo1_quality1.jpg", quality=1)

display(Image.open("demo1_quality95.jpg"))
display(Image.open("demo1_quality25.jpg"))
display(Image.open("demo1_quality1.jpg"))
```

```python
from PIL import Image, ImageDraw, ImageFont
from IPython.display import display

img = Image.open('demo2.jpg.jpeg')
I1 = ImageDraw.Draw(img)
font_path = "Roboto-Italic.ttf"
try:
    myFont = ImageFont.truetype(font_path, 95)
except IOError:
    myFont = ImageFont.load_default()

I1.text((20, 20), "beautiful pic", font=myFont, fill=(500, 0, 0))

display(img)
img.save("demo2_with_text.jpg")

```
